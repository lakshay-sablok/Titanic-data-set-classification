```python
X=df.drop(['Name','PassengerId','Survived','Ticket'], axis=1)
ticket=df["Ticket"]
y=df['Survived']
```

```python
X = pd.get_dummies(X, dtype=int)
df.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X['Ticket'] = le.fit_transform(ticket)
X.head()
```

```python
data=X
data["Survived"]=y
data.head()
```

```python
data.corr()
```

# **Features in descending order**

```python
corr_matrix = data.corr()

# Get the absolute correlation values with respect to the target variable
corr_with_target = abs(corr_matrix['Survived'])

# Sort the correlation values in descending order
sorted_corr= corr_with_target.sort_values(ascending=False)

print(sorted_corr)
```

# **Splitting the data into Test and Train**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=50)
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

# **Training the dataset with various algorithms and then testing and finding the accuracy **

```python
Accuracy = []
Recall = []
Precision = []
F1_Score = []
Algorithm = []
```

# **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score

logreg = LogisticRegression()

logreg.fit(X_train, y_train)

Algorithm.append('Logistic Regression')

y_train_pred = logreg.predict(X_train)
y_test_pred = logreg.predict(X_test)

train_accuracy = accuracy_score(y_train, y_train_pred)
train_precision = precision_score(y_train, y_train_pred)
train_recall = recall_score(y_train, y_train_pred)
train_f1 = f1_score(y_train, y_train_pred)

test_accuracy = accuracy_score(y_test, y_test_pred)
test_precision = precision_score(y_test, y_test_pred)
test_recall = recall_score(y_test, y_test_pred)
test_f1 = f1_score(y_test, y_test_pred)

print("Training Metrics:")
print("Accuracy:", train_accuracy)
print("Precision:", train_precision)
print("Recall:", train_recall)
print("F1 Score:", train_f1)

print("Testing Metrics:")
print("Accuracy:", test_accuracy)
print("Precision:", test_precision)
print("Recall:", test_recall)
print("F1 Score:", test_f1)

Accuracy.append(test_accuracy)
Recall.append(test_recall)
Precision.append(test_precision)
F1_Score.append(test_f1)
```

# Support Vector Classifier

```python
from sklearn.svm import SVC

svc = SVC()

svc.fit(X_train, y_train)

Algorithm.append('Support Vector Classifier')

y_train_pred = svc.predict(X_train)
y_test_pred = svc.predict(X_test)

train_accuracy = accuracy_score(y_train, y_train_pred)
train_precision = precision_score(y_train, y_train_pred)
train_recall = recall_score(y_train, y_train_pred)
train_f1 = f1_score(y_train, y_train_pred)

test_accuracy = accuracy_score(y_test, y_test_pred)
test_precision = precision_score(y_test, y_test_pred)
test_recall = recall_score(y_test, y_test_pred)
test_f1 = f1_score(y_test, y_test_pred)

print("Training Metrics:")
print("Accuracy:", train_accuracy)
print("Precision:", train_precision)
print("Recall:", train_recall)
print("F1 Score:", train_f1)

print("Testing Metrics:")
print("Accuracy:", test_accuracy)
print("Precision:", test_precision)
print("Recall:", test_recall)
print("F1 Score:", test_f1)

Accuracy.append(test_accuracy)
Recall.append(test_recall)
Precision.append(test_precision)
F1_Score.append(test_f1)
```

# KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

knn.fit(X_train, y_train)

Algorithm.append('KNeighbors Classifier')

y_train_pred = knn.predict(X_train)
y_test_pred = knn.predict(X_test)

train_accuracy = accuracy_score(y_train, y_train_pred)
train_precision = precision_score(y_train, y_train_pred)
train_recall = recall_score(y_train, y_train_pred)
train_f1 = f1_score(y_train, y_train_pred)

test_accuracy = accuracy_score(y_test, y_test_pred)
test_precision = precision_score(y_test, y_test_pred)
test_recall = recall_score(y_test, y_test_pred)
test_f1 = f1_score(y_test, y_test_pred)

print("Training Metrics:")
print("Accuracy:", train_accuracy)
print("Precision:", train_precision)
print("Recall:", train_recall)
print("F1 Score:", train_f1)

print("Testing Metrics:")
print("Accuracy:", test_accuracy)
print("Precision:", test_precision)
print("Recall:", test_recall)
print("F1 Score:", test_f1)

Accuracy.append(test_accuracy)
Recall.append(test_recall)
Precision.append(test_precision)
F1_Score.append(test_f1)
```
